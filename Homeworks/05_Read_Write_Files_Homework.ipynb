```python
directory = '/Users/eherre/Documents/Data/python_data_course/Python_fundamentals/sample_files/'
notes = directory + 'notes.txt'
with open(notes, 'r') as file:
    lines_number = 0
    words_number = 0
    for line in file:
        words = len(line.split())
        words_number += words
        lines_number += 1
    print(f"Number of lines: {lines_number}")
    print(f"Number of words: {words_number}")
```

```python
notes_copy = directory + 'notes_copy.txt'
with open(notes, 'r') as readfile:
    with open(notes_copy, 'w') as writefile:
        for line in readfile:
            writefile.write(line)

with open(notes_copy, 'r') as writefile:
    new_file = writefile.read()
    print(new_file)
```

```python
students = directory + 'students.csv'
honor_roll = directory + 'honor_roll.csv'

import csv

with open(students, 'r') as readfile:
    reader = csv.DictReader(readfile)

    with open(honor_roll, 'w') as writefile:
        writer = csv.DictWriter(writefile, fieldnames = reader.fieldnames)
        writer.writeheader()
        for r in reader:
            if float(r['grade']) >= 4.2:
                writer.writerow(r)

with open(students, 'r') as readfile:
    reader = csv.DictReader(readfile)
    print(list(reader))
with open(honor_roll, 'r') as readfile:
    reader = csv.DictReader(readfile)
    print(list(reader))
```

```python
import csv

grades_list = []
with open (students, 'r') as f:
    students_list = list(csv.DictReader(f))
    for r in students_list:
        grades_list.append(float(r['grade']))

avg_grade = sum(grades_list)/len(grades_list)
print(avg_grade)
        
```

```python
output = directory + 'output.txt'

with open(output, 'a+') as f:
    f.seek(0)
    run_number = 0
    for r in f:
        if r.startswith("Run #"):
            run_number += 1
    f.write(f"Run #{run_number + 1}\n")
    f.seek(0)
    print(f.read())
```

```python
data = directory + "data.json"
data_with_ts = directory + "data_with_ts.json"
import json

with open(data, 'r') as readfile:
    x = json.load(readfile)
    x["exported_at"] = "11:11"
    with open(data_with_ts, 'w') as writefile:
        json.dump(x, writefile, ensure_ascii = False, indent = 2)

with open(data_with_ts, 'r') as f:
    print(f.read())
```

```python
table = directory + "table.tsv"
table_converted = directory + "table_converted.csv"

with open(table, 'r', newline = "") as readfile:
    reader = csv.DictReader(readfile, delimiter = "\t")
    with open(table_converted, 'w') as writefile:
        writer = csv.DictWriter(writefile, fieldnames = reader.fieldnames)
        writer.writeheader()
        for r in reader:
            writer.writerow(r)
    
```

```python
def safe_read(path):
    try:
        with open(path, 'r') as readfile:
            return(readfile.read())
    except FileNotFoundError:
        return None

non_existing_file = directory + "non_existing.txt"
safe_read(non_existing_file)
```
