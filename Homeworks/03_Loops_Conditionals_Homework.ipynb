```python
consecutive_heads = []
total_flips = 0

while len(consecutive_heads) < 3:
    flip = random.randint(0,1)
    total_flips += 1

    if flip == 0:
        consecutive_heads.append(flip)
    else:
        consecutive_heads.clear()

print(total_flips)
print(consecutive_heads)
```

```python
nums = [3, 4, 10, 11, 2]
evens_sum = 0

for num in nums:
    if (num % 2) == 0:
        evens_sum += num

print ("The sum of evens is:", evens_sum)
```

```python
words = ["data", "engineering", "ai", "python"]
longest_word = ''

for i in range(len(words)):
    if len(words[i]) > len(longest_word):
        longest_word = words[i]

print (longest_word)
```

```python
word = "aabbbbccdd"
character_count = {}

for character in word:
    if character in character_count:
        character_count[character] += 1
    else:
        character_count[character] = 1

print(character_count)
```
