## 1. What pandas is and why it is useful

pandas is a Python library for working with structured data, especially tables.

It is useful because it makes common data tasks easier:

- loading data from files like CSVs
- exploring a dataset quickly
- selecting only the rows or columns you need
- filtering, sorting, and summarizing data

If Python lists and dictionaries are good for general programming, pandas is especially good for data analysis.

```python
import pandas as pd
```

## 2. Series vs DataFrame

The two main pandas objects are:

- `Series`: a one-dimensional labeled collection of data
- `DataFrame`: a two-dimensional table with rows and columns

A good way to think about it:

- a `Series` is similar to one column in a table
- a `DataFrame` is the whole table

```python
city_series = pd.Series(["Bogota", "Medellin", "Cali"], name="city")
```

```python
city_series
```

```python
student_df = pd.DataFrame(
    {
        "name": ["Ana", "Luis", "Sara"],
        "score": [88, 92, 85],
        "passed": [True, True, True],
    }
)

print("Series:\n")
print(city_series)
print("\nDataFrame:\n")
student_df
```

```python
student_df['name']
```

## 3. Rows, columns, indexes

In a pandas `DataFrame`:

- each horizontal entry is a **row**
- each vertical entry is a **column**
- the **index** labels the rows

The index starts at `0` by default, but pandas can also use other values as row labels.

```python
print("Row index:", list(student_df.index))
print("Columns:", list(student_df.columns))
print("\nRow with index 0:\n")
print(student_df.iloc[0,2])
```

```python
list(student_df.columns)
```

```python
print(student_df.loc[0,'name'])
```

## 4. Creating Series and DataFrames from Python data structures

One reason pandas feels approachable is that you can create tables directly from data structures you already know, like lists and dictionaries.

```python
# From a dictionary of lists
inventory_from_dict = pd.DataFrame(
    {
        "product": ["Notebook", "Mouse", "Keyboard"],
        "price": [4.5, 12.0, 25.0],
    }
)

# From a list of dictionaries
sales_from_records = pd.DataFrame(
    [
        {"month": "Jan", "sales": 120},
        {"month": "Feb", "sales": 150},
        {"month": "Mar", "sales": 165},
    ]
)

print("From a dictionary of lists:\n")
print(inventory_from_dict)
print("\nFrom a list of dictionaries:\n")
sales_from_records
```

```python
inventory_from_dict
```

## 5. A small toy dataset

Before reading a file, it is helpful to practice with a tiny dataset created directly in the notebook. This keeps the focus on pandas concepts.

```python
sales_df = pd.DataFrame(
    {
        "sale_id": [101, 102, 103, 104, 105, 106],
        "product": ["Laptop", "Mouse", "Desk", "Monitor", "Chair", "Keyboard"],
        "category": ["Electronics", "Electronics", "Furniture", "Electronics", "Furniture", "Electronics"],
        "price": [1200, 25.5, 300, 220, 150, 45],
        "quantity": [2, 10, 3, 4, 5, 7],
        "city": ["Bogota", "Cali", "Bogota", "Medellin", "Cali", "Bogota"],
        "in_stock": [True, True, False, True, True, True],
    }
)

sales_df
```

## 6. Basic data inspection

When you first open a dataset, you usually want quick answers to questions like:

- What does it look like?
- How many rows and columns does it have?
- What are the column names?
- What type of data is inside each column?

pandas gives us several simple tools for this.

```python
print("First 3 rows:\n")
print(sales_df.head(2))
print("\nLast 2 rows:\n")
sales_df.tail(2)
```

```python
print("Shape:", sales_df.shape)
print("Columns:", list(sales_df.columns))
print("Index:", list(sales_df.index))
print("\nData types:\n")
print(sales_df.dtypes)
print("\nInfo:\n")
sales_df.info()
```

```python
sales_df.describe(include="all")
```

```python
sales_df.nunique()
```

## 7. Selecting one column and multiple columns

Selecting data is one of the most common things you will do.

- Use `df["column_name"]` for one column
- Use `df[["col1", "col2"]]` for multiple columns

Notice that selecting one column returns a `Series`, while selecting multiple columns returns a `DataFrame`.

```python
print("One column (Series):\n")
print(sales_df["product"])
print("\nMultiple columns (DataFrame):\n")
sales_df[["product", "price", "quantity"]]
```

## 8. Basic filtering with simple conditions

Filtering means keeping only the rows that match a condition.

Examples:

- rows where `price > 30`
- rows where `city == "Bogota"`
- rows that match two conditions at the same time

```python
expensive_items = sales_df[sales_df["price"] > 30]
bogota_sales = sales_df[sales_df["city"] == "Bogota"]
electronics_in_stock = sales_df[(sales_df["category"] == "Electronics") & (sales_df["in_stock"])]

print("Rows where price > 30:\n")
print(expensive_items)
print("\nRows where city == 'Bogota':\n")
print(bogota_sales)
print("\nElectronics that are in stock:\n")
electronics_in_stock
```

## 9. Sorting values

Sorting helps you organize data so patterns are easier to spot.

Use `sort_values()` and choose:

- the column to sort by
- whether you want ascending or descending order

```python
sales_df.sort_values(["category","price"], ascending=False)
```

```python
print("Sorted by price (highest first):\n")
print(sales_df.sort_values("price", ascending=False))
print("\nSorted by category, then by price within each category:\n")
sales_df.sort_values(["category", "price"], ascending=[True, False])
```

## 10. Reading a CSV file with pandas

In real projects, your data will often come from a file. A very common format is CSV.

Below, we create a small local CSV file so the notebook runs anywhere. Later, you can replace the path with a Kaggle dataset or another CSV on your computer.

```python
csv_path = "sample_store_data.csv"  # Later, replace this with your own CSV path.

csv_example = pd.DataFrame(
    {
        "order_id": [201, 202, 203, 204, 205],
        "customer": ["Ana", "Luis", "Marta", "Diego", "Elena"],
        "product": ["Book", "Tablet", "Notebook", "Headphones", "Mouse"],
        "price": [18, 450, 6, 80, 20],
        "quantity": [2, 1, 5, 1, 3],
    }
)

# Save a local CSV so we can practice reading a file.
csv_example.to_csv(csv_path, index=False)

print(f"Sample CSV saved to: {csv_path}")
csv_example
```

```python
orders_df = pd.read_csv(csv_path)

print("Loaded from CSV:\n")
print(orders_df.head())
print("\nQuick inspection:")
print("Shape:", orders_df.shape)
print("Columns:", list(orders_df.columns))
```

## Mini challenge setup

Here is a fresh dataset for one final task.

```python
challenge_df = pd.DataFrame(
    {
        "student": ["Ana", "Luis", "Marta", "Jorge", "Sofia", "Camilo"],
        "hours_studied": [5, 2, 7, 4, 6, 3],
        "score": [78, 60, 91, 73, 88, 67],
        "passed": [True, False, True, True, True, False],
    }
)

challenge_df
```

## Mini challenge

Use `challenge_df` and do the following:

1. Inspect the dataset with at least three pandas tools such as `head()`, `shape`, `columns`, or `info()`.
2. Select only the `student` and `score` columns.
3. Filter the rows where `score >= 75`.
4. Sort the filtered result by `score` from highest to lowest.
5. Bonus: keep only students who passed **and** studied at least 5 hours.

```python
# Write your mini challenge solution here.

# Step 1: inspect the dataset

print(challenge_df.info(), "\n")
print(challenge_df.head(), "\n")
print(challenge_df.columns, "\n")

# Step 2: select columns

selection_df = challenge_df[["student","score"]]
print(selection_df, "\n")

# Step 3: filter rows

filtered_df = selection_df[selection_df["score"] >= 75]
print(filtered_df, "\n")

# Step 4: sort the result

filtered_df.sort_values(["score"], ascending = False)

# Bonus

filtered_bonus_df = challenge_df[(challenge_df["hours_studied"] >= 5) & (challenge_df["passed"] == True)]
filtered_bonus_df
```
